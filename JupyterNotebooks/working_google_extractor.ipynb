```python
import langextract as lx
import textwrap
import graph
import knowledge_graph





# Define the extraction task
prompt = textwrap.dedent("""\
    Extract characters, emotions, and relationships in order of appearance.
    Use exact text for extractions. Do not paraphrase or overlap entities.
    Provide meaningful attributes for each entity to add context.""")

prompt = knowledge_graph.system_prompt_1

# Provide a high-quality example
examples = [
    lx.data.ExampleData(
        text="""Adam is a software engineer in Microsoft since 2009, 
            "and last year he got an award as the Best Talent. Trump approved.""",
        extractions=[
            lx.data.Extraction(
                extraction_class="Person",
                extraction_text="Adam",
                attributes={"works_for": "Microsoft"}
            ),
            lx.data.Extraction(
                extraction_class="Person",
                extraction_text="Adam",
                attributes={"won_award": "Best Talent"}
            ),
            lx.data.Extraction(
                extraction_class="Person",
                extraction_text="Donald John Trump",
                attributes={"approved": "Adam"}
            )
        ]
    )
]

example2 = []
```

```python
# Simple extraction from a short text
input_text = '''Vincent de Groof (6 December 1830 – 9 July 1874) was a Dutch-born Belgian early pioneering aeronaut. He created an early model of an ornithopter and successfully demonstrated its use before fatally crashing the machine in London, UK.[1] '''

result = lx.extract(
      text_or_documents=input_text,
      prompt_description=prompt,
      examples=examples,
      model_id="llama3.2",
      model_mlflow_dburl="http://localhost:11434"
  )

# Display results
print(f"Extracted {len(result.extractions)} entities:\n")
for extraction in result.extractions:
    print(f"• {extraction.extraction_class}: '{extraction.extraction_text}'")
    if extraction.attributes:
        for key, value in extraction.attributes.items():
            print(f"  - {key}: {value}")
```
