# NER

## Initialize 

Initilize the Llm inference provider to get the entities from text.

```python
import sys
import os

sys.path.append(os.path.abspath("../src"))

from llm_ner_nel.inference_api.entity_inference import EntityInferenceProvider, display_entities
```

## Text --> Entities

```python
text = '''Vincent de Groof is a Dutch-born Belgian early pioneering aeronaut. He created an early model of an ornithopter'''

inference_provider = EntityInferenceProvider(model="llama3.2") 
entities = inference_provider.get_entities(text)

display_entities(entities, console_log=True)
```

```python
from transformers import AutoTokenizer, AutoModelForTokenClassification, pipeline

# Model fine-tuned for NER
#MODEL_NAME = "microsoft/deberta-v3-large"
MODEL_NAME = "tner/deberta-v3-large-ontonotes5"


tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME)
model = AutoModelForTokenClassification.from_pretrained(MODEL_NAME)

ner = pipeline(
    "ner",
    model=model,
    tokenizer=tokenizer,
    aggregation_strategy="simple",  # groups subword tokens into entities
    device=0
)

text = "Vincent de Groof is a Dutch-born Belgian early pioneering aeronaut. He created an early model of an ornithopter"

results = ner(text)

for entity in results:
    print(entity)
```

```python
from transformers import AutoTokenizer, AutoModelForTokenClassification
from transformers import pipeline
tokenizer = AutoTokenizer.from_pretrained("Davlan/xlm-roberta-large-ner-hrl")
model = AutoModelForTokenClassification.from_pretrained("Davlan/xlm-roberta-large-ner-hrl")
nlp = pipeline("ner", model=model, tokenizer=tokenizer)
example = "Vincent de Groof is a Dutch-born Belgian early pioneering aeronaut. He created an early model of an ornithopter"
ner_results = nlp(example)
print(ner_results)
```
